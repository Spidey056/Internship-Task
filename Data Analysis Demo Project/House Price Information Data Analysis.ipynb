```python
import re

def extract_and_replace_last_before(word):
    matches = re.findall(r'\b([^,]+)\s*,\s*[^,]+\b', word)
    return matches[-1] if matches else word

def replace_last_before_word(dataframe, column_name):
    dataframe[column_name] = dataframe[column_name].apply(extract_and_replace_last_before)

column_to_replace = 'Address'
replace_last_before_word(house_prices, column_to_replace)
```

```python
column_to_analyze = 'Address'

unique_words = house_prices[column_to_analyze].astype(str).str.split(',').explode().str.strip().unique()

print("Location of Houses :")
for i, word in enumerate(unique_words, start=1):
    print(f"{i}. {word}")
```

```python
pd.set_option('display.float_format', lambda x: '%.0f' % x)

# Your other code for searching and displaying rows
column_to_search = 'Address'  # Replace with the name of the column you want to search
name_to_search = input("Enter The Location:")  # Replace with the name you want to search

# Check for exact matches in a case-insensitive manner
matching_rows = house_prices[house_prices[column_to_search].astype(str).str.lower() == name_to_search.lower()]

# Reset the display float format option for other columns
pd.set_option('display.float_format', None)

# Convert matching rows to table format and print
if not matching_rows.empty:
    table = tabulate(matching_rows, headers='keys', tablefmt='pretty', showindex=False)
    print(table)
else:
    print(f"No matching rows found for '{name_to_search}' in '{column_to_search}'.")
```
